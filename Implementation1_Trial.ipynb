```python
#Seperate Features from response
def seperate(df_train):
    """
    input: dataframe
    """
    #Grab all continuous features
    df_train = df_train.drop(["yr_renovated","zipcode","yr_built","dummy","date"], axis=1)
    x = df_train.iloc[:,0:-1]
    
    #Grab response y
    y = df_train.iloc[:,-1]
    
    return x,y
    
    
#Add new features [Month, Day, Year]
def split_date(df_train):
    """
    splits date into seperate features
    input: dataframe
    """
    for i in range(0,len(df_train)):
        df_train.loc[i,"month"] = df_train.loc[i,"date"].split("/")[0]
        df_train.loc[i,"day"] = df_train.loc[i,"date"].split("/")[1]
        df_train.loc[i,"year"] = df_train.loc[i,"date"].split("/")[2]
    print("DONE")
    return df_train

#Normalize data
def normalize(df1):
    """
    Normalizes feature matrix
    input: feature df
    """
    x = (df1 - np.min(df1))/(np.max(df1) - np.min(df1))
    x["dummy"] = 1
    return x

#Linear regression function
def linear_regress(x,y,eta,t,lamb):
    """
    x: input/features
    y: opuput
    eta: learning rate
    t: iterations
    lamb: regularization constant
    """
    n = 0
    e = np.zeros(len(y))
    errors = np.zeros(t)
    gradient = np.zeros(t)
    
    #Initialize weights [w] and predictions [y_hat]
    w = np.zeros(len(x[0]))
    
     
    while n<t:
        #Initialize gradient for each epoch
        gradient_vector = np.zeros(len(x[0]))
            

        
        #Traverse through each data point   
        for i in range(len(x)):
            #Predicted value
            y_hat = np.dot(w.T,x[i])

            #Error
            e[i] = (y[i] - y_hat)**2 
            
            
            #Regularization
            if np.dot(w.T,w) == 0:
                r = 0
            else:
                r = (np.dot(w.T,w))**0.5
            
            #Traverse through each feature to update corresponding weights
            for j in range(len(x[0])):
                gradient_vector[j] += ((-2)*(y[i] - y_hat)*x[i,j]) + (2*lamb*r)
            
        w -=  eta*gradient_vector
        
        #print(gradient_vector)
        #print(np.dot(gradient_vector,gradient_vector)**0.5)
        errors[n] = 0.5*sum(e)
        gradient[n] = np.dot(gradient_vector.T,gradient_vector)**0.5
        
        #####
        if (n+1)%1 == 0:
            #print(f'WEIGHT : {w}')
            print(f'Epoch: {n+1}')
            print(f'Gradient: {gradient[n]}')
            print(f'Weight: {w}')
        #####
        n += 1
    
    return w,errors,gradient
```
